#Step 1: Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Step 2: Create the Data frame

```python
dataframe = pd.read_csv('Zomato Dataset.csv')
print(dataframe)
```

###Convert the Data type of Column - Rate

```python
def handlerate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

# Change 'handleRate' to 'handlerate' to match the function definition
dataframe['rate'] = dataframe['rate'].apply(handlerate)
print(dataframe.head())
```

##Checking Missing Value

```python
dataframe.info()
```

##**Type of Resturant**

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Resturant")
```

Conclusion: Majority of the resturant falls in Dinning Category


**How much vote has each type of restaurant received from customers**

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'Votes': grouped_data})
plt.plot(result, c="Blue", marker="o")
plt.xlabel("TYPE OF RESTURANT", c="black", size=10)
plt.ylabel("VOTES", c="Black", size=10)
```

Conclusion: Dinning resturant has recieved maximum votes

 **Rating received by majority of Restaurants**

```python
plt.hist(dataframe['rate'],bins =10)
plt.title("Rating Distribution")
plt.show()
```

Conclusion: the majority resturant received rating from 3.5 to 4


**Average spending on each order by Couples**

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Conclusion: the majority of couple prefer resturant with an approximate cost of 300 rupees

**Which mode has received the maximum rating(Online/Offline)**

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

Conclusion: Offline order received lower rating in comparison to online order

**Which types of restaurant received more offline orders, so that zomato can prefer customers with some good offers**

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("online_order") # Corrected the function name from xlable to xlabel
plt.ylabel("Listed In (Type)")
plt.show()
```

Conclusion: Dinning restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that cient prefer orders in person at restaurants, but prefer online ordering at cafes.
